```python
#looks like there are some zero values in the data we need to tackle
#Tme to clean the data
#lets start by checking for duplicates

df=df.drop_duplicates()


# check the shape after drop the duplicates
df.shape

```

```python
#no duplicates, so on to missing values
# Count of mull. values
# check the missing values in any column

# Display number of null values in every column in dataset

df.isnull().sum()
```

```python
#No null values, so let's check for zero values

print('No. of zero values in Glucose ',df[df['Glucose']==0].shape[0])
print('No. of zero values in BloodPressure',df[df['BloodPressure']==0].shape[0])
print('No. of zero values in SkinThickness ',df[df[ 'SkinThickness' ]==0].shape[0])
print('No. of zero values in Insulin',df[df[ 'Insulin']==0].shape[0])
print('No. of zero values in BMI',df[df['BMI' ]==0].shape[0])
```

```python
#Options:Diabetes to remove zeros or impute zeros with the mean, I'm going with option 2

#let's check glucose first before doing the remainder

df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
print('No. of zero values in Glucose' ,df[df[ 'Glucose' ]==0].shape[0])
```

```python
#it worked, now let's do the other variables

df[ 'BloodPressure']=df[ 'BloodPressure' ].replace(0,df['BloodPressure' ].mean())
df[ 'SkinThickness']=df[ 'SkinThickness' ].replace(0,df[ 'SkinThickness'].mean())
df['Insulin']=df['Insulin'].replace(0,df['Insulin' ].mean())
df['BMI']=df[ 'BMI' ].replace(0,df['BMI'].mean())
```

```python
#Outcome appears to be highly correlated with [preganancies, blood pressure, skin thickness, insulin, and diabetes pedigree function]
#let's select for these feature and split the data fram into x and y

target_name = 'Outcome'

# Separate object for target feature
y = df[target_name]

# Separate Object for Input Features
x = df.drop(target_name, axis=1)
x.head()
```

```python
y.head()
```

```python
# Apply Standard Scaler

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler. fit(x)

SSX = scaler. transform(x)
```

```python
#train and test the data, so must be split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(SSX, y, test_size=0.2, random_state=7)
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
#let's do logistic regression training

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='liblinear' ,multi_class='ovr')
lr.fit(x_train,y_train)
```

```python
## Making predictions 
lr_pred=lr.predict(x_test)
```

```python
lr_pred.shape
```

```python
#logistic regression model evaluation

# Train score & Test score of Logistic Regression
from sklearn.metrics import accuracy_score

print("Train Accuracy of Logistic Regression",lr.score(x_train, y_train)*100)
print("Accuracy (Test) score of Logistic Regression",lr.score(x_test, y_test)*100)
print("Accuracy (Test) score of Logistic Regression" ,accuracy_score(y_test,lr_pred)*100)
```

```python
#now let's try a decsion tree

from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
dt. fit(x_train,y_train)
```

```python
#Making predictions on test dataset
dt_pred=dt.predict(x_test)
dt_pred.shape
```

```python
# Train score & Test score of Decesion Tree

print("Train Accuracy of Decesion Tree", dt.score(x_train,y_train)*100)
print("Accuracy (Test) score of Decesion Tree",dt.score(x_test, y_test)*100)
print("Accuracy score of Decesion Tree",accuracy_score(y_test,dt_pred)*100)
```

```python
#now let's try a random forest

from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(criterion='entropy')
rf.fit(x_train,y_train)

RandomForestClassifier(criterion='entropy')
```

```python
#Making predictions on test dataset

rf_pred=rf.predict(x_test)
rf_pred.shape
```

```python
# Train score & Test score of Random Forest

print("Train Accuracy of Random Forest",rf.score(x_train,y_train)*100)
print("Accuracy (Test) score of Random Forest", rf.score(x_test, y_test)*100)
print("Accuracy score of Random Forest",accuracy_score(y_test,rf_pred)*100)
```

```python
# Making the Confusion Matrix of Logistic Regression Model

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
cm = confusion_matrix(y_test, lr_pred)

print('TN - True Negative {}'.format(cm[0,0]))

print('FP - False Positive {}'.format(cm[0,1]))

print('FN - False Negative {}'.format(cm[1,0]))
print('TP - True Positive {}'.format(cm[1,1]))

print('Accuracy Rate: {}'.format(np.divide(np.sum([cm[0,0],cm[1,1]]),np.sum(cm))*100))
print('Misclassification Rate: {}'.format(np.divide(np.sum([cm[0,1],cm[1,0]]),np.sum(cm))*100))
```
